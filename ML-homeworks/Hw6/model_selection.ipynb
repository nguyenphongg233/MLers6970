```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('kc_house_data.csv')
data.head(10)
data.drop(['id'], axis=1, inplace=True)
data.head(10)
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

scaler = StandardScaler()
feature = data['price']
X = data.drop(['price', 'date'], axis=1)

X_train,X_test,y_train,y_test = train_test_split(X, feature, test_size=0.2, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# linear regression
from sklearn.linear_model import LinearRegression
model_LR = LinearRegression()
model_LR.fit(X_train, y_train)
print("Linear Regression score: ", model_LR.score(X_test, y_test))
```

```python
# Lasso regression
from sklearn.linear_model import Lasso
model_Lasso = Lasso()
model_Lasso.fit(X_train, y_train)
print("Lasso Regression score: ", model_Lasso.score(X_test, y_test))
```

```python
# KNN regression
from sklearn.neighbors import KNeighborsRegressor
model_KNN = KNeighborsRegressor()
model_KNN.fit(X_train, y_train)
print("KNN Regression score: ", model_KNN.score(X_test, y_test))
```

```python
# Random Forest regression
from sklearn.ensemble import RandomForestRegressor
model_RF = RandomForestRegressor()
model_RF.fit(X_train, y_train)
print("Random Forest Regression score: ", model_RF.score(X_test, y_test))
```

```python
#Ridge regression
from sklearn.linear_model import Ridge
model_Ridge = Ridge()
model_Ridge.fit(X_train, y_train)
print("Ridge Regression score: ", model_Ridge.score(X_test, y_test))
```

```python
# Ridge regression with polynomial features
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
model_Ridge_poly = make_pipeline(PolynomialFeatures(degree=2), Ridge())
model_Ridge_poly.fit(X_train, y_train)
print("Ridge Regression with polynomial features score: ", model_Ridge_poly.score(X_test, y_test))
```
