```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# lay du lieu dataset - du lieu ve nha gia tu URL
# Note: load_boston was removed from sklearn 1.2+, so we load it from an external source
import urllib.request
import io

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]

print("Số chiều dữ liệu input: ", X.shape)
print("Số chiều dữ liệu target: ", y.shape)
print()
print("5 mẫu dữ liệu đầu tiên:")
print("input: ", X[:5])
print("target: ", y[:5])
```

```python
X_train = X[:404]
y_train = y[:404]

X_test = X[405:]
y_test = y[405:]

X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# library cho linear regression
from sklearn.linear_model import LinearRegression
import numpy as np

lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print("Root mean squared error: %.2f" % np.sqrt(mean_squared_error(y_test, y_pred)))
print("Coefficient of determination: %.2f" % r2_score(y_test, y_pred))

```

```python
# linear regression by hand 

A = np.hstack([X_train, np.ones((X_train.shape[0], 1))])  # Thêm cột bias
w = np.linalg.inv(A.T @ A) @ A.T @ y_train  # Tính toán trọng số w
A_test = np.hstack([X_test, np.ones((X_test.shape[0], 1))])  # Thêm cột bias cho tập test
y_pred_hand = A_test @ w  # Dự đoán bằng cách nhân ma trận

# RMSE cho linear regression by hand
print("Root mean squared error (by hand): %.2f" % np.sqrt(mean_squared_error(y_test, y_pred_hand)))
```
